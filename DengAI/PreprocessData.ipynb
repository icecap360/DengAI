```
library(tidyverse)
#install.packages("imputeTS")
library(imputeTS)
getwd()
dat <- read.csv('dengue_features_train.csv')
labels <- read.csv('dengue_labels_train.csv')
test <- read.csv('dengue_features_test.csv')
dat <- dat %>% mutate(total_cases = labels$total_cases, isTrain = 1.0)

#as always I will combine the train and test data to ensure the same preprocessing happens on both
#This will also facilitate better imputations that because the test data starts right after the train data 
#ends
dat <- rbind(dat, test %>% mutate(total_cases=-1, isTrain=0.0))
dat$reanalysis_sat_precip_amt_mm
```

```
nrow(dat)
#na is not a serious problem, other then ndvi_ne and ndvi_nw
#they both also have small variance
#week_start_date should really be a date instead
data.frame(colnames = names(dat), classes=sapply(dat, class), nas=colSums(is.na(dat)), 
  variance=sapply(dat, function(x) {ifelse(is.numeric(x), sd(x, na.rm = T), NA)}), 
  range=sapply(dat, function(x) {ifelse(is.numeric(x), range(x, na.rm = T), NA)}) )

```

```
#normalize each column with mean and standard deviation from training set
nums <- dat %>% select_if(is.numeric) %>% select(-total_cases) %>% mutate_all(as.double)
factors <- dat %>% select_if(function(x) {!is.numeric(x)}) %>% mutate(total_cases = dat$total_cases)
nums.mean <- colMeans(nums %>% filter(isTrain == 1), na.rm=TRUE)
nums.sd <-    apply(nums      %>% filter(isTrain == 1), 2, sd, na.rm=TRUE)
nums <- sweep(nums, 2, nums.mean)
nums <- sweep(nums, 2, nums.sd, FUN="/")
nums$isTrain <- dat$isTrain #restore the isTrain column because  nan's have appeared
dat <- cbind(nums, factors)
```

```
#A NAIVE dataset
#Impute the other NA's means of previous two weeks and next two weeks data in the same city

fill.na.kalman <- function(d) {
  d %>% mutate(week_start_date = as.Date(as.character(week_start_date)))  %>% 
    arrange(week_start_date) %>% group_by(city) %>% mutate_all(na_kalman)
}
fill.na.exp <- function(d) {
  d %>% mutate(week_start_date = as.Date(as.character(week_start_date)))  %>% 
    arrange(week_start_date) %>% group_by(city) %>% mutate_all(na_ma)
}
fill.na.spline <- function(dat) {
  dat %>% mutate(week_start_date = as.Date(as.character(week_start_date)))  %>% 
    arrange(week_start_date) %>% group_by(city) %>% mutate_all(function(x) {na_interpolation(x,option="spline")})
}
naive_dat <- fill.na.kalman(dat)


#OBTAIN TEST AND TRAIN DATA FROM THE COMBINE DATAFRAME
get.train <- function(d) {
  d %>% filter(isTrain == 1) %>% select(-isTrain)
}
get.test <- function(d) {
  d %>% filter(isTrain == 0) %>% select(-total_cases, -isTrain)
}
naive_dat_train <- get.train(naive_dat)
write.csv(naive_dat_train , "sample_data/naive_data_train.csv")
naive_data_test <- get.test(naive_dat)
write.csv(naive_data_test , "sample_data/naive_data_test.csv")
```

The best solution I can think of is impute the missing values, this is because of the time nature series of the data. However we should make a binary variable for na's in feature columns with lots of na's (like ndvi_ne). 
 We should also consider other imputation techniques, like smooth spline and exponential moving average

```
#I will assume NA percipitation means ZERO percipitation 
new.firsthalf.fill.na <- function(d) {
  d %>% mutate_at(c("reanalysis_sat_precip_amt_mm", "reanalysis_precip_amt_kg_per_m2", 
    "station_precip_mm", "precipitation_amt_mm"), function(x) {ifelse(is.na(x) , 0,x)}) %>% 
    filter(!is.na(reanalysis_air_temp_k) || !isTrain)
  
}
#there seem to be two clusters
#ones that have na in Normalized difference vegetation index 
#second is examples that have na in data from the NOAA's GHCN daily climate database
tdat <- new.firsthalf.fill.na(dat)
num.na <- tdat %>% is.na()  %>% rowSums() 
tdat[num.na>=3,]
```

```
new.secondhalf.fill.na <- function (d ) {
  d %>% mutate(isna_ndvi_ne = ifelse(is.na(ndvi_ne), 1, 0 ), 
                isna_ndvi_nw = ifelse(is.na(ndvi_nw), 1, 0 ),
                isna_station_avg_temp_c = ifelse(is.na(station_avg_temp_c), 1,0), 
                isna_station_diur_temp_rng_c = ifelse(is.na(station_diur_temp_rng_c), 1,0)
   )
}
ndat <- new.secondhalf.fill.na(tdat)
#Kalman Imputation
ndat1 <- fill.na.kalman(ndat)
ndat1_train <- get.train(ndat1)
write.csv(ndat1_train , "sample_data/kalman_data_train.csv")
ndat1_test <- get.test(ndat1)
write.csv(ndat1_test , "sample_data/kalman_data_test.csv")

#Exponential Moving Average Imputation
ndat2 <- fill.na.exp(ndat)
ndat2_train <- get.train(ndat2)
write.csv(ndat2_train , "sample_data/exp_data_train.csv")
ndat2_test <- get.test(ndat2)
write.csv(ndat2_test , "sample_data/exp_data_test.csv")

#Spline Interpolation Imputation
ndat3 <- fill.na.exp(ndat)
ndat3_train <- get.train(ndat3)
write.csv(ndat3_train , "sample_data/spline_data_train.csv")
ndat3_test <- get.test(ndat3)
write.csv(ndat3_test , "sample_data/spline_data_test.csv")

```
